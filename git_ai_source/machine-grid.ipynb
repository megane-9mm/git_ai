```python
from sklearn.datasets import load_iris

#irisデータのロード
iris = load_iris()

#ロードしたデータを、特徴量とラベルに分ける
X = iris.data
y = iris.target
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
#グリッドサーチを実施
param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features':[2, 4, 6, 8]},
    {'bootstrap':[False], 'n_estimators':[3, 10], 'max_features':[2, 3, 4]}
]

forest_reg = RandomForestRegressor()

grid_serch = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')

grid_serch.fit(X, y)
```

```python
grid_serch.best_params_
```

GridSearchCVを用いることで、グリッドサーチと交差検証を交えた検証を行うことができる。  
一つのパラメータで複数回グリッドサーチを行い、最良のパラメータを求めてみる。

```python
#グリッドサーチを実施
#n_estimators(作成木数)
param_grid = [
    {'n_estimators': [3, 10, 30, 50, 75, 100, 200, 300, 400, 500]}
]

forest_reg = RandomForestRegressor()

grid_serch = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')

grid_serch.fit(X, y)
```

```python
grid_serch.best_params_
```

```python
cvres = grid_serch.cv_results_
for mean_score, params in zip(cvres['mean_test_score'], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

グリッドサーチより、最良のn_estimatorsは200だと分かった。  
また、スコアは”0.2377368993937065 {'n_estimators': 200}”である。

```python
#グリッドサーチを実施
param_grid = [
    {'n_estimators': [3, 10, 30, 50, 75, 100, 200], 'max_depth':[2, 4, 6, 8, 10], 'max_features':[2, 3, 4]}
]

forest_reg = RandomForestRegressor(random_state=20200923)

grid_serch = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')

grid_serch.fit(X, y)
```

単体で試した時と結果が異なっている。

```python
# 基本処理ライブラリ
import pandas as pd
import numpy as np

# データ可視化ライブラリ
import matplotlib.pyplot as plt
%matplotlib inline

# 機械学習ライブラリ
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
# 特徴量作成
x = np.array([20, 30, 100, 120, 220, 240])
x = x[:,None].astype('float64')

# ターゲット作成
y = np.array([10, 30, 120, 110, 80, 70])
y = y[:,None].astype('float64')
```

```python
# 特等量とターゲットを可視化
plt.scatter(x,y, marker='o')
plt.xlabel("Advertisement Spend")
plt.ylabel("Sales")
plt.show()
```

```python
# ステージ1 推測値（yの平均）
y_mean = y.mean()
y_mean = np.full((6,1),y_mean)
M1_pred = y_mean.copy()
M1_pred
```

```python
# ステージ1 残差（実際の値 - 推測値）
error_1 = y.reshape(6,1) - y_mean
error_1
```

```python
# ステージ1の確認
table = pd.DataFrame({
    '広告費':x[:,0],
    '売上': y[:,0],
    'M1推測': M1_pred[:,0],
    'M1残差': error_1[:,0]},
    index=["A","B","C","D","E","F"]
)
# テーブル表示
table[['広告費', '売上', 'M1推測', 'M1残差']]
```

```python
# ステージ1の結果を可視化
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 6))

# 予測のプロット
ax1.plot(x, y, 'o', label='Actual Values')
ax1.plot(x, M1_pred, 'r', label='Prediction')
ax1.set_title('Decision Tree: 1st Stage')
ax1.set_xlabel('Advertising Spend')
ax1.set_ylabel('Profit')
ax1.legend(fontsize=8)

# 誤差のプロット
ax2.plot(x, error_1, 'o', color='purple',label='Residual Values')
ax2.set_xlim(0,250)
ax2.set_ylim(-80,80)
ax2.hlines(y=0, xmin=0, xmax=300, color='g', alpha=0.5, label='Perfect Model')
ax2.set_title('Residuals')
ax2.set_xlabel('Advertising Spend')
ax2.set_ylabel('Residuals')
ax2.legend(fontsize=8)
```

```python
# ステージ2のモデル構築（M_2）
M2 = DecisionTreeRegressor(max_depth=1)
M2 = M2.fit(x, error_1)
M2_output = M2.predict(x)
```

```python
# データ解析のライブラリ
import pandas as pd 
import numpy as np

#　データ可視化のライブラリ
import matplotlib.pyplot as plt

# 機械学習ライブラリ
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score

# XGBoost
import xgboost as xgb 
from xgboost import XGBClassifier 
```

```python
# データセットの読み込み
iris = datasets.load_iris()
type(iris)
```

```python
# ターゲットのクラス名を確認
print(iris.target_names)

# ターゲットの値を確認
print(set(iris.target))
```

```python
irisdf = pd.DataFrame(iris.data, columns = iris.feature_names)
irisdf['target'] = iris.target
irisdf.head()
```

```python
# 訓練データとテストデータの分割
train_set, test_set = train_test_split(irisdf, test_size = 0.2, random_state = 42)
print(train_set.shape)
print(test_set.shape)
```

```python
# 特徴量（x）とターゲット（y）に分割
X_train = train_set.drop('target',axis=1)
y_train = train_set['target'].copy()

X_test = test_set.drop('target',axis=1)
y_test = test_set['target'].copy()
```

```python
# 訓練/テストデータをNumpy配列からDMatrixへ変換
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
print(type(dtrain))
print(type(dtest))
```

```python
# 訓練データ（DMatrix）の名前を確認
dtrain.feature_names
```

```python
# 訓練データ（DMatrix）の特徴量のデータ型を確認
print(dtrain.feature_types)
print(dtest.feature_types)
```

```python
# XGBoostのハイパーパラメータ 
param = {
    'max_depth': 3,
    'eta': 0.3,
    'silent': 0,
    'objective': 'multi:softprob',
    'num_class': 3
}
```

```python
# ブースティング反復回数の設定
num_boost_round = 20
```

```python
# XGBoost モデル訓練
xgboost_model = xgb.train(param, dtrain, num_boost_round)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#プロット範囲を決定
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.1), np.arange(x2_min, x2_max, 0.1))

#プロットを描く枠を用意
f, ax = plt.subplots(nrows=1, ncols=2)

#それぞれのモデルの結果を描く
for idx, clf, tt in zip([0, 1], [gbdt, gbdt], ['Decision tree', 'Bagging']):
    #モデルの学習と予測
    clf.fit(X_train, y_train)
    Z = clf.predict(np.array([xx1.ravel(), xx2.ravel()]).T) #ravel()は多次元のリストを1次元に変換
    Z = Z.reshape(xx1.shape)
    #分類境界を描く
    ax[idx].contourf(xx1, xx2, Z, alpha=0.3)
    #各ラベルのサンプルを描く
    ax[idx].scatter(x=X_train[y_train==0, 0], y=X_train[y_train==0, 1], c='blue', marker='^')
    ax[idx].scatter(x=X_train[y_train==1, 0], y=X_train[y_train==1, 1], c='green', marker='o')
    ax[idx].scatter(x=X_train[y_train==2, 0], y=X_train[y_train==2, 1], c='red', marker='x')
    #タイトル
    ax[idx].set_title(tt)

#軸ラベル
ax[0].set_ylabel('od280/od315_of_diluted_wines', fontsize=12)
ax[0].set_xlabel('alcohol', fontsize=12)

plt.tight_layout()
plt.show()
```
